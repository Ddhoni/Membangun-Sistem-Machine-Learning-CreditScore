### Ganti Simbol aneh dengan NaN

```python
def text_cleaning(val):
    # Jika nilainya NaN → biarkan tetap NaN
    if pd.isna(val):
        return np.nan
    # Jika string → bersihkan simbol-simbol tertentu
    if isinstance(val, str):
        return val.strip('_ ,"')
    # Jika bukan string (angka, float, dsb) → biarkan apa adanya
    return val

# --- Terapkan hanya ke kolom bertipe object/string ---
obj_cols = data.select_dtypes(include='object').columns

data[obj_cols] = data[obj_cols].apply(lambda col: col.map(text_cleaning))

# --- Ganti string aneh atau kosong dengan NaN ---
data[obj_cols] = data[obj_cols].replace(
    ['', 'nan', '!@9#%8', '#F%$D@*&8'],
    np.nan
)

data.head()
```

```python
def clean_and_transform(data: pd.DataFrame) -> pd.DataFrame:
    """
    Membersihkan dan mengonversi tipe data untuk kolom-kolom numerik & string tertentu.
    """
    data= data.copy()

    # ID & Customer_ID dari hex → int
    data['ID'] = data['ID'].apply(lambda x: int(x, 16) if isinstance(x, str) else np.nan)
    data['Customer_ID'] = data['Customer_ID'].apply(
        lambda x: int(x[4:], 16) if isinstance(x, str) and len(x) > 4 else np.nan
    )

    # Month "January" → 1..12
    data['Month'] = pd.to_datetime(data['Month'], format='%B', errors='coerce').dt.month

    # Age (nullable integer)
    data['Age'] = pd.to_numeric(data['Age'], errors='coerce').astype('Int64')

    # SSN: hapus '-' lalu ke float; tangani NaN
    data['SSN'] = data['SSN'].apply(
        lambda x: np.nan if pd.isna(x)
        else float(str(x).replace('-', '')) if isinstance(x, (str, int, float)) else np.nan
    )

    # Kolom numerik lain
    num_cols = [
        'Annual_Income', 'Num_of_Loan', 'Num_of_Delayed_Payment',
        'Changed_Credit_Limit', 'Outstanding_Debt',
        'Amount_invested_monthly', 'Monthly_Balance'
    ]
    for col in num_cols:
        data[col] = pd.to_numeric(data[col], errors='coerce')

    return data

data = clean_and_transform(data)
data
```

```python
def preprocess_credit_and_loan(data: pd.DataFrame):
    data_out = data.copy()

    # 1) Konversi umur kredit ke bulan
    def Month_Converter(x):
        if pd.isna(x):
            return np.nan
        s = str(x).lower()
        try:
            s = s.replace('years','year').replace('months','month')
            parts = s.split()
            nums = [int(tok) for tok in parts if tok.isdigit()]
            if len(nums) >= 2:
                years, months = nums[0], nums[1]
                return float(years * 12 + months)
            if len(nums) == 1:
                return float(nums[0])
        except Exception:
            pass
        return np.nan

    if 'Credit_History_Age' in data_out.columns:
        data_out['Credit_History_Age'] = (
            data_out['Credit_History_Age'].apply(Month_Converter).astype(float)
        )

    # 2) Normalisasi Type_of_Loan
    if 'Type_of_Loan' in data_out.columns:
        data_out['Type_of_Loan'] = data_out['Type_of_Loan'].apply(
            lambda x: x.lower().replace('and ', '').replace(', ', ',').strip()
            if pd.notna(x) else x
        )

    # 3) Ekstraksi nilai unik Type_of_Loan — versi SESUAI KODE LAMA
    def get_Diff_Values_Colum(data_column, diff_value=[], sep=',', replace=''):
        column = data_column.dropna()
        for i in column:
            if sep not in i and i not in diff_value:
                diff_value.append(i)
            else:
                for data in map(lambda x:x.strip(), re.sub(replace, '', i).split(sep)):
                    if not data in diff_value:
                        diff_value.append(data)
        return dict(enumerate(sorted(diff_value)))

    loan_dict = {}
    if 'Type_of_Loan' in data_out.columns:
        loan_dict = get_Diff_Values_Colum(data_out['Type_of_Loan'])

    # Output tambahan (jaga kompatibel dengan rangkaian explorasi kamu)
    cha_by_customer_list = None
    tol_value_counts_head20 = None
    tol_counts_by_customer = None
    tol_by_customer_list = None

    if {'Customer_ID','Credit_History_Age'}.issubset(data_out.columns):
        cha_by_customer_list = data_out.groupby('Customer_ID')['Credit_History_Age'].apply(list)

    if 'Type_of_Loan' in data_out.columns:
        tol_value_counts_head20 = data_out['Type_of_Loan'].value_counts(dropna=False).head(20)
    if {'Customer_ID','Type_of_Loan'}.issubset(data_out.columns):
        tol_counts_by_customer = data_out.groupby('Customer_ID')['Type_of_Loan'].value_counts(dropna=False)
        tol_by_customer_list = data_out.groupby('Customer_ID')['Type_of_Loan'].apply(list)

    return (
        data_out,
        loan_dict,
        cha_by_customer_list,
        tol_value_counts_head20,
        tol_counts_by_customer,
        tol_by_customer_list,
    )
(
    data_processed,
    loan_dict,
    cha_by_cust_list,
    tol_vc_head20,
    tol_counts_by_cust,
    tol_by_cust_list
) = preprocess_credit_and_loan(data)

print("✅ Selesai preprocess.\n")
print("👉 data_processed.head():")
display(data_processed.head())

print("\n👉 Jumlah baris/kolom:", data_processed.shape)

if loan_dict:
    print("\n👉 loan_dict (contoh 20 teratas):")
    first_20 = dict(list(loan_dict.items())[:20])
    print(first_20)

if cha_by_cust_list is not None:
    print("\n👉 Credit_History_Age list per Customer_ID (head):")
    display(cha_by_cust_list.head())

if tol_vc_head20 is not None:
    print("\n👉 Type_of_Loan value_counts (top 20):")
    display(tol_vc_head20)

if tol_counts_by_cust is not None:
    print("\n👉 Type_of_Loan value_counts per Customer_ID (head):")
    display(tol_counts_by_cust.head())

if tol_by_cust_list is not None:
    print("\n👉 Type_of_Loan list per Customer_ID (head):")
    display(tol_by_cust_list.head())

# Jika ingin mengganti `data` asli:
data = data_processed
```

A. Object Column NaN Values: Reassign Group Mode Values¶


```python
import numpy as np
import pandas as pd

# =============================================
# 1️⃣ FUNGSI: Isi NaN kolom object per grup dengan modus (mode)
# =============================================
def fill_object_nan_by_group_mode(
    data: pd.DataFrame,
    group_col: str,
    columns: list,
    *,
    inplace: bool = True,
    verbose: bool = True
) -> pd.DataFrame:
    """
    Mengisi nilai NaN/None pada kolom object per 'group_col'
    dengan modus (mode) dari grup tsb.

    Args:
        data        : DataFrame input
        group_col  : kolom pengelompok (mis. 'Customer_ID')
        columns    : daftar kolom object yang ingin dibersihkan
        inplace    : apakah hasil langsung ditulis ke data
        verbose    : tampilkan log sebelum/sesudah

    Returns:
        DataFrame hasil bersih
    """
    work = data if inplace else data.copy()

    for col in columns:
        if col not in work.columns:
            if verbose:
                print(f"⚠️ Kolom '{col}' tidak ditemukan — dilewati.")
            continue

        # Ganti None -> np.nan
        mask_none = work[col].isin([None])
        if mask_none.any():
            work.loc[mask_none, col] = np.nan

        # Info sebelum
        if verbose:
            n_nan_before = work[col].isna().sum()
            print(f"\n[{col}] NaN sebelum: {n_nan_before}")
            try:
                sample_before = work.groupby(group_col)[col].apply(list).head()
                print(f"[{col}] Contoh sebelum (5 grup):")
                for lst in sample_before.values:
                    print(lst)
            except Exception:
                pass

        # Hitung mode per grup
        mode_per_group = (
            work.groupby(group_col)[col]
                .transform(lambda s: s.mode().iloc[0] if not s.mode().empty else np.nan)
        )

        # Isi nilai kosong dengan mode
        work[col] = work[col].fillna(mode_per_group)

        # Info sesudah
        if verbose:
            n_nan_after = work[col].isna().sum()
            print(f"[{col}] NaN sesudah: {n_nan_after}")
            try:
                sample_after = work.groupby(group_col)[col].apply(list).head()
                print(f"[{col}] Contoh sesudah (5 grup):")
                for lst in sample_after.values:
                    print(lst)
            except Exception:
                pass

    return work


# =============================================
# 2️⃣ IMPLEMENTASI LANGSUNG KE DATA
# =============================================

# Ganti nilai NaN di Type_of_Loan jadi "No Data" (sesuai instruksi kamu)
if 'Type_of_Loan' in data.columns:
    data['Type_of_Loan'] = data['Type_of_Loan'].fillna('No Data')

# Jalankan pengisian NaN pada kolom-kolom object
data = fill_object_nan_by_group_mode(
    data,
    group_col='Customer_ID',
    columns=[
        'Name',
        'Occupation',
        'Credit_Mix',
        'Payment_Behaviour'
    ],
    inplace=True,
    verbose=True
)

# =============================================
# 3️⃣ TAMPILKAN HASIL
# =============================================

print("\n✅ Selesai mengisi NaN kolom object per grup dengan modus.")
print("👉 Data setelah pembersihan:")
display(data.head(10))  # tampilkan 10 baris pertama

print("\n📊 Ringkasan jumlah NaN per kolom:")
display(data.isna().sum().to_frame('Jumlah_NaN'))

print("\n📋 Tipe data setelah pembersihan:")
display(data.dtypes.to_frame('dtype'))
```

```python
import numpy as np
import pandas as pd

def Numeric_Wrong_Values_Reassign_Group_Min_Max(data, groupby, column, inplace=True):
    """
    Bersihkan nilai numerik di 'column' per 'groupby':
      - Tentukan range min–max (berdasarkan mode per-grup → fallback quantile 5–95 → min/max global → 0/0)
      - Nilai di luar range -> NaN
      - Imput NaN: mode per-grup -> median per-grup -> median global -> 0
    """
    work = data if inplace else data.copy()
    work[column] = pd.to_numeric(work[column], errors='coerce')

    # 1) Hitung range min–max dari mode per-grup
    cur = (
        work.loc[work[column].notna()]
            .groupby(groupby)[column]
            .apply(list)
    )
    if len(cur) > 0:
        # ambil satu nilai mode per grup (jika multi-mode, ambil yang pertama)
        mode_vals = []
        for vals in cur:
            m = pd.Series(vals).mode()
            if not m.empty:
                mode_vals.append(m.iloc[0])
        if len(mode_vals) > 0:
            mini, maxi = float(np.min(mode_vals)), float(np.max(mode_vals))
        else:
            mini = maxi = np.nan
    else:
        mini = maxi = np.nan

    # Fallback range jika belum valid
    s_non_na = work[column].dropna()
    if (not np.isfinite(mini)) or (not np.isfinite(maxi)):
        if not s_non_na.empty:
            mini = float(s_non_na.quantile(0.05))
            maxi = float(s_non_na.quantile(0.95))
        else:
            mini = maxi = 0.0
    if mini > maxi and not s_non_na.empty:
        mini, maxi = float(s_non_na.min()), float(s_non_na.max())

    print(f"\n[{column}] Range min–max yang digunakan: ({mini}, {maxi})")

    # 2) Tandai out-of-range menjadi NaN
    marked = work[column].where(~((work[column] < mini) | (work[column] > maxi)), np.nan)

    # 3) Imput NaN berjenjang
    # 3.1 mode per-grup
    mode_by_group = work.groupby(groupby)[column].transform(
        lambda x: x.mode()[0] if not x.mode().empty else np.nan
    )
    result = marked.fillna(mode_by_group)

    # 3.2 median per-grup
    if result.isna().any():
        median_by_group = work.groupby(groupby)[column].transform(lambda x: x.median())
        result = result.fillna(median_by_group)

    # 3.3 median global
    if result.isna().any():
        global_med = work[column].median()
        if pd.isna(global_med):
            global_med = 0.0
        result = result.fillna(global_med)

    # 3.4 sisa → 0
    result = result.fillna(0.0)

    # 4) Assign
    if inplace:
        work[column] = result
        print(f"✅ {column} selesai diproses. Sisa NaN: {work[column].isna().sum()}")
        return work
    else:
        print(f"✅ {column} selesai diproses (return Series).")
        return result


def clean_numeric_columns(data, group_col='Customer_ID', inplace=True):
    """
    Membersihkan kolom numerik umum + interpolasi Credit_History_Age per group.
    Menggunakan Numeric_Wrong_Values_Reassign_Group_Min_Max untuk tiap kolom.
    """
    numeric_cols = [
        'Age',
        'SSN',
        'Annual_Income',
        'Monthly_Inhand_Salary',
        'Num_Bank_Accounts',
        'Num_Credit_Card',
        'Interest_Rate',
        'Num_of_Loan',
        'Delay_from_due_date',
        'Num_of_Delayed_Payment',
        'Changed_Credit_Limit',
        'Num_Credit_Inquiries',
        'Outstanding_Debt',
        'Total_EMI_per_month',
        'Amount_invested_monthly',
        'Monthly_Balance'
    ]

    work = data if inplace else data.copy()

    for col in numeric_cols:
        if col in work.columns:
            print(f"\n==================== Cleaning: {col} ====================")
            Numeric_Wrong_Values_Reassign_Group_Min_Max(
                work, group_col, col, inplace=True
            )

    # Interpolasi Credit_History_Age per group (gunakan transform agar index align)
    if 'Credit_History_Age' in work.columns:
        print("\nInterpolating Credit_History_Age ...")
        work['Credit_History_Age'] = (
            pd.to_numeric(work['Credit_History_Age'], errors='coerce')
              .groupby(work[group_col])
              .transform(lambda s: s.interpolate().bfill().ffill())
        )

    print("\n✅ Semua kolom numerik selesai dibersihkan.")
    return work
```

### Fixing Anomalies

- Num_Bank_Accounts,
- Delay_from_due_date,
- Num_of_Delayed_Payment,
- Monthly_Balance have negative strange number
- Amount_invested_monthly have positive strange number

```python
import numpy as np
import pandas as pd

def fix_strange_values(data: pd.DataFrame, group_col: str = 'Customer_ID', inplace: bool = True) -> pd.DataFrame:
    """
    Perbaiki nilai 'strange' sesuai instruksi:
      - Num_Bank_Accounts < 0 -> 0
      - Delay_from_due_date < 0 -> NaN -> perbaiki dengan Numeric_Wrong_Values_Reassign_Group_Min_Max
      - Num_of_Delayed_Payment < 0 -> NaN -> perbaiki dengan Numeric_Wrong_Values_Reassign_Group_Min_Max
      - Monthly_Balance < 0 -> NaN -> perbaiki dengan Numeric_Wrong_Values_Reassign_Group_Min_Max
      - Amount_invested_monthly >= 10000 -> NaN -> isi mode per group
    """
    work = data if inplace else data.copy()

    # Num_Bank_Accounts
    if 'Num_Bank_Accounts' in work.columns:
        work.loc[work['Num_Bank_Accounts'] < 0, 'Num_Bank_Accounts'] = 0
        print("✅ Num_Bank_Accounts negatif diganti 0")

    # Delay_from_due_date
    if 'Delay_from_due_date' in work.columns:
        work.loc[work['Delay_from_due_date'] < 0, 'Delay_from_due_date'] = np.nan
        Numeric_Wrong_Values_Reassign_Group_Min_Max(work, group_col, 'Delay_from_due_date', inplace=True)
        print("✅ Delay_from_due_date negatif dibersihkan & diisi ulang (per group)")

    # Num_of_Delayed_Payment
    if 'Num_of_Delayed_Payment' in work.columns:
        work.loc[work['Num_of_Delayed_Payment'] < 0, 'Num_of_Delayed_Payment'] = np.nan
        Numeric_Wrong_Values_Reassign_Group_Min_Max(work, group_col, 'Num_of_Delayed_Payment', inplace=True)
        print("✅ Num_of_Delayed_Payment negatif dibersihkan & diisi ulang (per group)")

    # Monthly_Balance
    if 'Monthly_Balance' in work.columns:
        work.loc[work['Monthly_Balance'] < 0, 'Monthly_Balance'] = np.nan
        Numeric_Wrong_Values_Reassign_Group_Min_Max(work, group_col, 'Monthly_Balance', inplace=True)
        print("✅ Monthly_Balance negatif dibersihkan & diisi ulang (per group)")

    # Amount_invested_monthly
    if 'Amount_invested_monthly' in work.columns:
        work.loc[work['Amount_invested_monthly'] >= 10000, 'Amount_invested_monthly'] = np.nan
        work['Amount_invested_monthly'] = work.groupby(group_col)['Amount_invested_monthly'].transform(
            lambda x: x.mode()[0] if not x.mode().empty else np.nan
        )
        print("✅ Amount_invested_monthly outlier >10000 dibersihkan & diisi mode per group")

    return work


# ======== IMPLEMENTASI LANGSUNG ========

data = fix_strange_values(data, group_col='Customer_ID', inplace=True)

print("\n✅ Semua nilai aneh sudah diperbaiki.")
display(data.head(10))
```

## Assign Categorical Types to Numeric Types


```python
def encode_categorical_columns(data: pd.DataFrame, columns: list = None, suffix: str = "_Num") -> pd.DataFrame:
    """
    Mengonversi kolom kategorikal menjadi kode numerik (category codes).
    - Secara default mengenali kolom kategorikal umum dari dataset kredit.
    - Otomatis buat kolom baru dengan suffix (default: '_Num').
    """
    work = data.copy()

    # Kolom default (bisa disesuaikan)
    if columns is None:
        columns = [
            'Occupation', 'Credit_Mix',
            'Payment_of_Min_Amount', 'Payment_Behaviour'
        ]

    for col in columns:
        if col in work.columns:
            new_col = f"{col}{suffix}"
            work[new_col] = (
                work[col]
                .astype('category')
                .cat.codes
                .replace(-1, np.nan)   # -1 artinya NaN di pandas category
            )
        else:
            print(f"Kolom '{col}' tidak ditemukan, dilewati.")

    return work

data = encode_categorical_columns(data)
data
```
